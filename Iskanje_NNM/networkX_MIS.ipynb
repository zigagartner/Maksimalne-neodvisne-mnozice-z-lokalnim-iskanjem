```python
import networkx as nx

G = nx.hypercube_graph(6)
nx.draw(G)
V = list(G.nodes)
V[10]
```

```python
X = nx.maximal_independent_set(G)
print(X)
```

```python
colour_map = []
for v in range(len(V)):
    if V[v] in X:
        colour_map.append("green")
    else:
        colour_map.append("blue")
        
nx.draw(G, node_color=colour_map)
```
